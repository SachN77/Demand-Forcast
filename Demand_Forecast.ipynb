```python
import pandas as pd
import numpy as np

df = pd.read_csv('Historical Product Demand.csv')
df
```

```python
#Checking for missing values in the DataFrame
df.isnull().sum()
```

```python
# Removed the Date column because of missing values
#df.drop('Date', axis = 1, inplace = True)
df.dropna(inplace=True,axis=0)
df
```

```python
print(len(df['Product_Code']))
print(len(df['Warehouse']))
print(len(df['Product_Category']))
print(len(df['Order_Demand']))
result = df.dtypes
result
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
forcast_df = pd.get_dummies(df, columns=["Product_Code", "Warehouse", "Product_Category"])

# df1 = ohe.fit_transform(df[['Product_Code', 'Warehouse', 'Product_Category']])
# pd.DataFrame(df1.toarray(), columns=ohe.categories_).head()
forcast_df
```

```python

forcast_df['Order_Demand'] = pd.to_numeric(forcast_df['Order_Demand'], errors='coerce')
#print(forcast_df)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

forcast_df['Order_Demand'] = scaler.fit_transform(np.array(forcast_df['Order_Demand']).reshape(-1,1))
forcast_df=forcast_df.replace(r'^\s*$', np.nan, regex=True)
forcast_df.dropna(inplace=True,axis=0)


forcast_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_columns = 2198
X = forcast_df.iloc[:2000, 2:]
y = forcast_df.iloc[:2000, 1]
train_ratio = 0.80
test_ratio = 0.20
validation_ratio = 0.5

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_ratio)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = validation_ratio)
```

```python
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor(criterion='mse', random_state=0)
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)
tree_score = r2_score(y_test, y_pred)
print(tree_score)
```

```python
from sklearn.svm import SVR
modelSVR = SVR(kernel='poly').fit(X_train, y_train)
y_pred_svr = modelSVR.predict(X_test)
score_svr = r2_score(y_test, y_pred_svr)
score_svr
```
